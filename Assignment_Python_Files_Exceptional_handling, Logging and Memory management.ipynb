```python
#Q9.What is the difference between multithreading and 
#multiprocessing?
```

```python
#Multithreading vs. Multiprocessing:

#Multithreading and multiprocessing are two different approaches
#to achive concurency in programming.

#Multithreading: 

#Definition: Multithreading is a technique where multiple threads
#are created within a single proess, shaing the the same memory space.

#Characteristics: 

#-Threads share the same memory space.

#-Threads are lightweight and have a lower overhead compared to
#processes.

#-Threads can communicate with each other using shared variables or queues.

#Use Cases:

#I/O-boud tasks, such as reading/writing files or network 
#communication.
#GUI applications where responsiveness is important.

#Multiprocessing:

#Definition: Multiprocessing is a technique where multiple 
#processes are created, each with its own memory space.

#Characteristics:

#-Processes have their own memory space.

#-Processes are heavier and have a higher overhead compared to
#threads.

#-Processes can communicate with each other using inter-process
#communication (IPC) mechanisms like pipes, queues, or shared memory.

#Use Cases:
#-CPU-bound tasks, such as scientific computing, data processing,
#or machine learning.

#-Tasks that require true parallelism and can take advantage of 
#multiple CPU cores.

#Key Differences:

#Memory Space: Multithreading shares the same memory space, while
#multiprocessing has separate memory spaces for each process.

#Overhead: Multithreading has a lower overhead compared to 
#multiprocessing.

#Parrallelism: Multprocessing can achieve true parrellelism on 
#mult-core CPUs, while multithreading is limited bu the Global 
#Interpreter Lock (GIL) in Python.

#Choosing Between Multithreading and Multiprocesing:

#I/O-bound tasks: Multithreading is often a good choice for I/O-boud tasks.

#CPU-bound tasks: Multiprocessing is often a good choice for 
#CPU-bound tasks that require true parallelism.

#Example:

import threading
import multiprocessing
import time

def worker(num):
    print(f"Worker {num} started")
    time.sleep(2)
    print(f"Worker {num} finished")

threads = []
for i in range(5):
    t = threading.Thread(target=worker, args=(i,))
    threads.append(t)
    t.start()

processes = []
for i in range(5):
    p = multiprocessing.Process(target=worker, args=(i,))
    processes.append(p)
    p.start()
```

```python
#Ans15:

#Python Garbage Collection System: Python's garbage collection 
#system is a memory managment machanism that automatically frees 
#up memory occupied by objects that are no longer needed or 
#referenced.

#Reference Counting:

#1.Object Creation: When an object is created, Python assigns 
#a reference count of 1 to it.

#2.Refernce Increment: When the object is assigned to a new variable 
#or stored in a data structure, its reference count is inremented

#3.Reference Decrement: When the object is no longer referenced, its 
#reference count is decremented.

#4.Object Deletion: When the reference count reaches 0, the 
#object is deleted, and its memory is freed.

#Generational Garbage Collection:

#1.Generations: Python divides objects into three generations
#base on their lifetime:

#Generation 0: Newly created objects.
#Generation 1: Objects that survive one garbage collection cycle.
#Generation 2: Long lived objects.

#2.Garbage Collection Cycles: Python periodically runs garbage
#collection cycles to identify and free up memory ocupied by 
#unreachable objects.

#Benefits:

#1.Memory Management: Python's garbage collection system 
#automatically manages memory, reducing the risk of memory 
#leaks and dangling pointers.

#2.Efficient Memory Use: Garbage collection helps optimize 
#memory use by freeing up memory occupied by objects that 
#are no longer needed.

#Best Practices:

#1.Avoid Circular References: Avoid creating circular references,
#which can prevent objects from being garbage collected.

#2.Use Weak References: Use weak reference to allow objects to 
#be garbage collected while still maintaining a reference to them.
```

```python
#Q4.Write a Python script that reads from one file and writes
#its content to another file
```

```python
#Ans4:

def copy_file(source_filename, destination_filename):
    try:
        with open(source_filename, 'r') as source_file:
            contents = source_file.read()
        with open(destination_filename, 'w') as destination_file:
            destination_file.write(contents)
        print(f"File '{source_filename}' copied to '{destination_filename}' successfully.")
    except FileNotFoundError:
        print(f"File '{source_filename}' not found.")
    except Exception as e:
        print(f"An error occurred; {e}")

copy_file('example3.txt','example4.txt')
```

```python
#Q18.How would you implement a basic logging setup that logs 
#to a file with rotation after 1MB?
```

```python
#Ans18:

import logging
from logging.handlers import RotatingFileHandler

logger = logging.getLogger(__name__)
logger.setLevel(logging.INFO)

handler = RotatingFileHandler('app.log', maxBytes=1024*1024, backupCount=5)
handler.setLevel(logging.INFO)

formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)
logger.addHandler(handler)

logger.info('This is an info message.')
logger.warning('This is a warning message.')
logger.error('This is an error message.')
```

```python
#Q21. Write a Python program that reads a file and prints
#the number of occurrences of a specific word.
```

```python
#Ans21:

def count_word_occurrences(filename, word):
    try:
        with open(filename, 'r') as file:
            contents = file.read().lower().split()
            word_count = contents.count(word.lower())
            return word_count
    except FileNotFoundError:
        print(f"The file '{filename}' was not found.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

def main():
    filename = input("Enter the filename:")
    word = input("Enter the word to search for:")
    word_count = count_word_occurrences(filename, word)
    if word_count is not None:
        print(f"The word '{word}' occurs {word_count} times in the file.")

if __name__ == "__main__":
    main()
```

```python
#Q23. Write a Python program that writes to a log file 
#when an error occurs during file handling.
```

```python
#Ans23:

import logging

logging.basicConfig(filename ='file_handling_errors.log', level=logging.ERROR)

def handle_file(filename):
    try:
        with open(filename, 'r') as file:
            contents = file.read()
            print(contents)
    except FileNotFoundError:
        logging.error(f"File'{filename}' not found.")
    except PermissionError:
        logging.error(f"Permission denied to access file '{filename}'.")
    except Exception as e:
        logging.error(f"An error occurred while handling file '{filename}': {e}")

def main():
    filename = 'example2.txt'
    handle_file(filename)
if __name__ == "__main__":
    main()
```
